```python
def explore_database(db_file):
    """
    Connects to the SQLite database and lists all tables and their columns.

    Parameters:
    db_file (str): Path to the SQLite database file.
    """
    # Connect to the SQLite database
    con = sqlite3.connect(db_file)
    cur = con.cursor()
    
    # List all tables
    cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cur.fetchall()
    print("Tables in the database:", tables)
    
    # List columns in each table
    for table in tables:
        table_name = table[0]
        cur.execute(f"PRAGMA table_info({table_name});")
        columns = cur.fetchall()
        print(f"Columns in {table_name}:", columns)
    
    # Close the connection
    con.close()

# Define the database file path
DB_FILE = '/local/DSPT/data/nlp-data.db'  # Update this path to your database file

# Explore the database
explore_database(DB_FILE)
```

```python
def load_data(db_file, query):
    """
    Connects to the SQLite database, executes the given query, and returns a DataFrame.

    Parameters:
    db_file (str): Path to the SQLite database file.
    query (str): SQL query to execute.

    Returns:
    pd.DataFrame: DataFrame containing the query results.
    """
    # Connect to the SQLite database
    con = sqlite3.connect(db_file)
    cur = con.cursor()

    # Execute the query and fetch all results
    cur.execute(query)
    data = cur.fetchall()

    # Get column names from the cursor
    column_names = [description[0] for description in cur.description]

    # Create a DataFrame from the fetched data
    df = pd.DataFrame(data, columns=column_names)

    # Close the connection
    con.close()

    return df

def save_to_csv(df, file_name):
    """
    Saves the DataFrame to a CSV file.

    Parameters:
    df (pd.DataFrame): DataFrame to save.
    file_name (str): Path to the output CSV file.
    """
    df.to_csv(file_name, index=False)
    print(f'Data saved to {file_name}')

# Define a general SQL query to retrieve a large dataset
QUERY = """
    SELECT 
        Labels.NumericValue AS label,
        RawTexts.Text AS text, 
        RawTexts.LengthCharacters AS length_characters, 
        RawTexts.HasEmoji AS has_emoji
    FROM Documents
    INNER JOIN RawTexts ON Documents.RawTextID = RawTexts.RawTextID
    INNER JOIN Labels ON Documents.LabelID = Labels.LabelID
    WHERE Documents.Type = "Review"
    ORDER BY RANDOM() LIMIT 5000
"""

# Load the data from the database
sample = load_data(DB_FILE, QUERY)

# Save the data to a CSV file
save_to_csv(sample, 'review_sample.csv')
```

## Extracting Data

```python
# Connect to the database
con = sqlite3.connect(DB_FILE)
cur = con.cursor()

# Define the SQL query
QUERY = """
    SELECT 
        Documents.DocumentID, 
        RawTexts.Text, 
        Labels.NumericValue AS Label, 
        Labels.Type AS LabelType, 
        Labels.StringValue AS LabelName,  
        Documents.Type AS DocumentType,
        RawTexts.LengthCharacters,
        RawTexts.HasEmoji
    FROM Documents
    INNER JOIN RawTexts ON Documents.RawTextID = RawTexts.RawTextID
    INNER JOIN Labels ON Documents.LabelID = Labels.LabelID
    ORDER BY RANDOM() LIMIT 1000;
"""

# Execute the query and load data into a DataFrame
reviews = pd.read_sql_query(QUERY, con)
con.close()
```

```python
reviews
```

## clean text

```python
# Function to clean text
def clean_text(text):
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

# Clean the text data
reviews['cleaned_text'] = reviews['Text'].apply(clean_text)

# Tokenize the text
reviews['tokens'] = reviews['cleaned_text'].apply(word_tokenize)

# Remove stop words
stop_words = set(stopwords.words('english'))
reviews['tokens'] = reviews['tokens'].apply(lambda x: [word for word in x if word not in stop_words])

# Display the cleaned and tokenized data
reviews[['cleaned_text', 'tokens']]
```

```python
# Flatten the list of tokens for frequency analysis
all_words = [word for tokens in reviews['tokens'] for word in tokens]

# Frequency distribution
freq_dist = nltk.FreqDist(all_words)

# Display the most common words
print(freq_dist.most_common(30))
```

```python
# Create a bigram and trigram frequency distribution
vectorizer = CountVectorizer(ngram_range=(2, 3))
X = vectorizer.fit_transform(reviews['cleaned_text'])
n_grams = vectorizer.get_feature_names_out()
n_grams_counts = X.toarray().sum(axis=0)

# Display the most common n-grams
n_grams_freq = pd.DataFrame({'ngram': n_grams, 'count': n_grams_counts})
print(n_grams_freq.sort_values(by='count', ascending=False).head(20))
```

```python
# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(all_words))

# Plot word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Rank and frequency
freqs = list(freq_dist.values())
ranks = np.arange(1, len(freqs) + 1)
indices = np.argsort(-np.array(freqs))
frequencies = np.array(freqs)[indices]

# Plotting Zipf's law
plt.figure(figsize=(10, 6))
plt.loglog(ranks, frequencies, marker=".")
plt.title("Zipf's Law")
plt.xlabel("Rank")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
